```python
import numpy as np
import matplotlib.pyplot as plt
import cechmate as cm
from persim import plot_diagrams
```

```python
filtration = []

# Add 0-simplices
filtration.append(([0],1))
filtration.append(([1],2))
filtration.append(([2],3))
filtration.append(([3],4))

# Add 1-simplices
m = 5
for i in range(4):
    for j in range(4):
        if i < j:
            filtration.append(([i,j],m))
            m += 1

# Add 2-simplices
for i in range(4):
    for j in range(4):
        for k in range(4):
            if i < j and j < k:
                filtration.append(([i,j,k],m))
                m += 1

# Add Final 3-simplex
filtration.append(([0,1,2,3],45))
```

```python
#Compute persistence diagrams
dgms = cm.phat_diagrams(filtration, show_inf = True)
print("H0:\n", dgms[0])
print("H1:\n", dgms[1])
print("H1:\n", dgms[2])
```

```python
plot_diagrams(dgms)
plt.show()
```
